```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Excess_Deaths_Associated_with_COVID-19.csv")
```

```python
df['Week Ending Date'] =  pd.to_datetime(df['Week Ending Date'])
df.info()
```

```python
df['State'].unique()
```

```python
indexNames = df[ (df['State'] == 'Puerto Rico') | 
                 (df['State'] == 'District of Columbia') |
                 (df['State'] == 'United States') |
                 (df['State'] == 'New York City')].index

df.drop(indexNames , inplace=True) 
df['State'].nunique()
```

```python
plot_test = df[(df['State'] == 'New York') & 
               (df['Week Ending Date'] > '2020-01-30') & 
               (df['Week Ending Date'] < '2020-07-01') &
               (df['Type'] == 'Predicted (weighted)') & 
               (df['Outcome'] == 'All causes')]
plot_test.plot.line(x='Week Ending Date', y=['Average Expected Count','Observed Number','Excess Higher Estimate'])
```

```python
plot_test.plot.line(x='Week Ending Date', y='Percent Excess Higher Estimate')
```

```python
plot_test.groupby('State')['Percent Excess Higher Estimate'].mean()
```

```python
group_by_state = df[(df['Week Ending Date'] > '2020-04-30') & 
                    (df['Week Ending Date'] < '2020-06-01') &
                    (df['Type'] == 'Predicted (weighted)') & 
                    (df['Outcome'] == 'All causes')].groupby('State')['Percent Excess Higher Estimate'].mean()
group_by_state
```

```python
daily_cases_deaths = pd.read_csv('/Users/robbdunlap/Google Drive/Work/COVID/Data Experiments/covid_app/covid_app/data/daily_cases_deaths.csv')
daily_cases_deaths['Date'] =  pd.to_datetime(daily_cases_deaths['Date'])
daily_cases_deaths.drop( daily_cases_deaths[ daily_cases_deaths['Province_State'] == 'District of Columbia' ].index , inplace=True)
list_of_states = daily_cases_deaths['Province_State'].unique().tolist()

daily_cases_deaths.drop(['new_cases_daily_diff','new_deaths_daily_diff',
                         'new_cases_rollavg_daily_diff','new_deaths_rollavg_daily_diff',],
                        axis=1, inplace=True)
daily_cases_deaths.shape
```

```python
adjusted_deaths = daily_cases_deaths.copy()
adjusted_deaths['adusted_date'] = adjusted_deaths['Date'] - timedelta(days=14)
```

```python
adjusted_deaths.shape
```

```python
def correctedInfections(row):
    for state in list_of_states:
        if row["Province_State"] == state:
            return (group_by_state[state] * row['new_deaths_rollavg']) + row['new_deaths_rollavg']
            


adjusted_deaths["corrected_deaths"] = adjusted_deaths.apply(correctedInfections, axis=1)
```

del adjusted_deaths
