# Best day of the week to grocery shop in FFX Co, Arlington, and Alexandria

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gmd = pd.read_csv("data/Global_Mobility_Report.csv")
```

```python
gmd.head()
```

```python
grocs = gmd[(gmd['sub_region_1'] == 'Virginia') &
           ((gmd['sub_region_2'] == 'Alexandria') |
            (gmd['sub_region_2'] == 'Arlington County')| 
            (gmd['sub_region_2'] == 'Fairfax County')) & 
            (gmd['date'] > '2020-06-30') ].copy()

grocs['date'] = pd.to_datetime(grocs['date'])

grocs
```

```python
grocs2 = grocs[['date','grocery_and_pharmacy_percent_change_from_baseline']].copy()
```

```python
grocs2.tail()
```

```python
grocs2['date'] = pd.to_datetime(grocs2['date'])
```

```python
grocs2.set_index(['date'], inplace=True)
grocs2.plot(legend=False)
```

```python
days_of_week = grocs2.groupby(grocs2.index.strftime('%A')).mean()
```

```python
days_ordered = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
days_of_week = days_of_week.reindex(days_ordered)
```

```python
days_of_week.plot.bar(legend=False)
```

```python
days_of_week.head()
```

```python
days_of_week = days_of_week['grocery_and_pharmacy_percent_change_from_baseline'] + 100
```

```python
ax = days_of_week.plot.bar(legend=False)
ax.set_ylabel('% Activity relative to Normal')
```
